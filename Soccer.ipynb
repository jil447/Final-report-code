```python
import matplotlib.pyplot as plt
import numpy as np
 

x = [2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]
y1 = [46,28,43,26,37,41,51,28,38,11]
y2 = [14,9,22,24,13,19,24,20,16,13]
y3 = [34,31,47,35,25,33,27,37,35,24]
y4 = [23,31,16,40,29,31,25,21,21,13]
# Source from Hupu App
        

plt.plot(x, y1, 'yo-', label="Messi")
plt.plot(x, y2, 'ro-', label="Neymar")
plt.plot(x, y3, 'ko-', label="C.Ronaldo")
plt.plot(x, y4, 'go-', label="L.Suarez")


for a, b1 in zip(x, y1):
    plt.text(a, b1, str(b1), ha="center", va="bottom", fontsize=8) 
for a, b2 in zip(x, y2):
    plt.text(a, b2, str(b2), ha="center", va="bottom", fontsize=8)  
for a, b3 in zip(x, y3):
    plt.text(a, b3, str(b3), ha="center", va="bottom", fontsize=8)
for a, b4 in zip(x, y4):
    plt.text(a, b4, str(b4), ha="center", va="bottom", fontsize=8)

plt.legend(loc = "upper right")

plt.title("2012-2021 Four Most Popular Soccer Players' Goals")
plt.xlabel("Career Year")
plt.ylabel("Number Of Goals")



plt.grid(color = 'olivedrab', linestyle = '--', linewidth = 0.25)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

barWidth = 0.1
fig = plt.subplots(figsize =(12, 8))

y1 = [46,28,43,26,37,41,51,28,38,11]
y2 = [14,9,22,24,13,19,24,20,16,13]
y3 = [34,31,47,35,25,33,27,37,35,24]
y4 = [23,31,16,40,29,31,25,21,21,13]
# Source from Hupu App

br1 = np.arange(len(y1))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]

plt.bar(br1, y1, color ='y', width = barWidth, label ='Messi')
plt.bar(br2, y2, color ='r', width = barWidth, label ='Neymar')
plt.bar(br3, y3, color ='b', width = barWidth, label ='C.Ronaldo')
plt.bar(br4, y4, color ='g', width = barWidth, label ='L.Suarez')


plt.title("2012-2021 Four Most Popular Soccer Players' Goals")
plt.xlabel("Career Year")
plt.ylabel("Number Of Goals")

plt.xticks([i + barWidth for i in range(len(y1))],
		['2012', '2013', '2014','2015', '2016', '2017', '2018', '2019','2020', '2021'])

plt.legend()


plt.grid(color = 'olivedrab', linestyle = '--', linewidth = 0.25)
plt.show()

```
